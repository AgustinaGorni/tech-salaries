2. Printear porcentaje de participación por región, ordenado de mayor a menor. Ej: 

- Ciudad Autónoma de Buenos Aires - 59.4%

````python
import collections

counter_regions = collections.Counter(regions)

#print(counter_regions)
participation_regions = {}

#Calcular porcentaje
def percent(value,total):
    """
    calculates a percentage and rounds to two decimals
    """   
    return round(float(value)/float(total) *100, 2)

for region in counter_regions:
    participation = percent(counter_regions[region],len(regions))
    participation_regions[region] = participation
    participation_regions = dict(sorted(participation_regions.items(), key=lambda item: item[1], reverse=True))
````

8. Formatear las carreras universitarias:
- Nombres capitalizados
- Reemplazar vocales con tilde por vocales sin tilde.
- Reemplazar `ñ` por `n`
- *lic*, *lic.* por Licenciatura
- *tec*, *tec.* por Tecnucatura
* *cs, *cs.* por Ciencias
* *ed, ed.* por Educación
* Transformaciones que se consideren necesarias

Printear porcentaje según carrera

````python
#Función para formatear los strings

def format_str(word:str):
    '''
    Removes accent marks, capitalizes and removes abbreviations
    ''' 
    word = word.lower().replace('.', '').replace(',', '')
    replace = (('á', 'a'), ('é', 'e'), 
               ('í', 'i'), ('ó', 'o'), 
               ('ú', 'u'), ('ñ', 'n'),
               ('lic ', 'licenciatura'), ('tec ', 'tecnicatura'), 
               ('cs ', 'ciencias'), ('ed ', 'educacion'))
    for a, b in replace:
        word = word.replace(a , b)
    
    word = word.capitalize()

    return word
````

````python
#MEDIANA
import statistics

salary_region = {}
salary_role = {}
salary_experience = {}
salary_level = {}
salary_career = {}
salary_technology = {}
salary_language = {}
salary_gender = {}
#Esta condición es para descartar a los que pusieron 1 en el campo de "Salario Neto"
for answer in all_answers:
    if answer['Salario dolares'] > max(dollar_values): 
        salary_region.setdefault(answer['Ciudad'], []).append(answer['Salario dolares'])
        salary_role.setdefault(answer['Puesto de trabajo'], []).append(answer['Salario dolares'])
        salary_experience.setdefault(answer['Años de experiencia'], []).append(answer['Salario dolares'])
        salary_level.setdefault(answer['Nivel de estudios alcanzado'], []).append(answer['Salario dolares'])
        salary_career.setdefault(answer['Carrera'], []).append(answer['Salario dolares'])
        salary_technology.setdefault(answer['Plataformas'], []).append(answer['Salario dolares'])
        salary_language.setdefault(answer['Lenguajes de programación o tecnologías'], []).append(answer['Salario dolares'])
        salary_gender.setdefault(answer['Me identifico'], []).append(answer['Salario dolares'])


def median_list (list_name):
    
    import statistics

    for key in list_name:
        med_value = statistics.median(list_name[key])
        list_name[key] = med_value

    return list_name

salary_region = median_list(salary_region)
salary_role = median_list(salary_role)
salary_role = dict(sorted(salary_role.items(), key=lambda item: item[1], reverse=True))
salary_experience = median_list(salary_experience)
salary_level = median_list(salary_level)
salary_career = median_list(salary_career)
salary_technology = median_list(salary_technology)
salary_language = median_list(salary_language)
salary_gender = median_list(salary_gender)

all_salaries = []
for answer in all_answers:
    if answer['Salario dolares'] != 0.0:
        all_salaries.append(answer['Salario dolares'])

salary_med = statistics.median(all_salaries)
````
